## Q1 - The equations of motion of the quarter-car are given by:

Car Body: $ \quad m_1 \ddot{x_1} = - 2 k_1 (x_1 - x_2) - c (\dot{x_1} - \dot{x_2}) $

Tire: $ \quad m_2 \ddot{x_2} = - k_2 (x_2 - p(t)) + 2 k_1 (x_1 - x_2) + c (\dot{x_1} - \dot{x_2}) $

## Q2 - Simulation of the quarter-car 

````python
import numpy as np
from casadi import *
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define system parameters
m1 = 40  # kg
m2 = 5   # kg
k1 = 9000   # N/m
k2 = 15000  # N/m
c_std = 200  # Ns/m for Q2

# Skyhook control parameters
c_min = 0  # Ns/m
c_max = 2000  # Ns/m

# Define the step disturbance
def p(t, step_size):
    return step_size if t >= 0 else 0

# Define the control law for c(t)
def semiactive_suspension(x1_dot, x2_dot, K):
    if x1_dot * (x1_dot - x2_dot) > 0:
        c = K * x1_dot / (x1_dot - x2_dot)
        c = np.clip(c, c_min, c_max)
    else:
        c = c_min
    return c

# Define the equations of motion
def equations(t, y, K, damper, step_size, skyhook=False):
    x1, x2, x1_dot, x2_dot = y
    if skyhook:
        c = semiactive_suspension(x1_dot, x2_dot, K)
    else:
        c = damper
    x1_ddot = (-k1 * (x1 - x2) - c * (x1_dot - x2_dot)) / m1
    x2_ddot = (-k2 * (x2 - p(t, step_size)) + k1 * (x1 - x2) + c * (x1_dot - x2_dot)) / m2
    return [x1_dot, x2_dot, x1_ddot, x2_ddot]
````

````python
def Q2(c=c_std, step_size=0.05):
    # Initial conditions: [x1, x2, x1_dot, x2_dot]
    y0 = [0, 0, 0, 0]

    # Time span for the simulation
    t_span = (0, 5)  # 5 seconds
    t_eval = np.linspace(t_span[0], t_span[1], 500)

    # Solve the differential equations
    sol = solve_ivp(equations, t_span, y0, t_eval=t_eval, args=(0, c_std, step_size, False))

    # Extract the results
    x1 = sol.y[0]
    x2 = sol.y[1]
    t = sol.t

    # Plot the results
    plt.figure(figsize=(10, 6))
    plt.plot(t, x1, label='Car Body Displacement $x_1(t)$')
    plt.plot(t, x2, label='Tire Displacement $x_2(t)$')
    plt.plot(t, [p(_, step_size) for _ in t], label='Step Disturbance $p(t)$', linestyle='--', color='green')
    plt.xlabel('Time [s]')
    plt.ylabel('Displacements [m]')
    plt.title(f'Simulation - {c:.3f} Ns/m Damping')
    plt.legend()
    plt.grid()
    #filename = f"./images/Q2.png"
    #plt.savefig(filename, dpi=700, format='png', bbox_inches='tight')
    plt.show()

# Run Q2
Q2()
````

## Q3 - Implementation of the skyhook control policy

````python
def Q3(K, step_size=0.05):
    # Initial conditions: [x1, x2, x1_dot, x2_dot]
    y0 = [0, 0, 0, 0]

    # Time span for the simulation
    t_span = (0, 5)  # 5 seconds
    t_eval = np.linspace(t_span[0], t_span[1], 500)

    # Solve the system of differential equations
    sol = solve_ivp(equations, t_span, y0, t_eval=t_eval, args=(K, c_std, step_size, True))

    # Extract the results
    x1 = sol.y[0]
    x2 = sol.y[1]
    t = sol.t

    # Plot the results
    plt.figure(figsize=(10, 6))
    plt.plot(t, x1, label='Car Body Displacement $x_1(t)$')
    plt.plot(t, x2, label='Tire Displacement $x_2(t)$')
    plt.plot(t, [p(_, step_size) for _ in t], label='Step Disturbance $p(t)$', linestyle='--', color='green')
    plt.xlabel('Time [s]')
    plt.ylabel('Displacements [m]')
    plt.title(f'Simulation with Semi-active Suspension - K = {K:.3f}')
    plt.legend()
    plt.grid()
    filename = f"./images/Q3-d={step_size}.png"
    plt.savefig(filename, dpi=700, format='png', bbox_inches='tight')
    plt.show()

# Run Q3 with a different values of K
K = [500, 1000, 2000]
for k in K:
    Q3(k)
````

## Q4 & Q5 - Formulation of the optimization problems and solutions for three road disturbances

### Optimal passive suspension

##### Objective Function:

\begin{aligned}
&\text{minimize} && J(c) = \int_{0}^{T} [x_1(t) - x_{1,\text{ss}}]^2 \, dt \\
&\text{subject to} && c \geq 0 \\
&&& c \leq c_{\text{max}}
\end{aligned}


where:
- $J(c)$ is the cost function to be minimized
-  $x_1(t)$  is the car body displacement at time \( t \),
- $ x_{1,\text{ss}} $ is the steady-state displacement of the car body,
- $ T $ is the total time of simulation,
- $ c_{\text{max}} $ is the maximum allowable damping coefficient.


````python
def Q4_1(step_size, check_convexity=False):
    T = 5  # Total simulation time

    # Define the optimization problem
    opti = Opti()

    # Decision variable and constraints (within reasonable bounds)
    c = opti.variable(1, 1)
    opti.subject_to(c_min <= c)
    opti.subject_to(c <= c_max)

    # Set an initial guess for c
    initial_guess = 1000
    opti.set_initial(c, initial_guess)

    # Time and state variables
    t = MX.sym('t')
    x = MX.sym('x', 4, 1)
    x1, x2, x1_dot, x2_dot = x[0], x[1], x[2], x[3]

    # Equations of motion
    x1_ddot = (-k1 * (x1 - x2) - c * (x1_dot - x2_dot)) / m1
    x2_ddot = (-k2 * (x2 - step_size) + k1 * (x1 - x2) + c * (x1_dot - x2_dot)) / m2
    rhs = vertcat(x1_dot, x2_dot, x1_ddot, x2_ddot)

    # Define the ODE function
    ode = {'x': x, 'p': vertcat(c, t), 'ode': rhs}
    F = integrator('F', 'cvodes', ode, {'tf': T/500})

    # Initial conditions
    x0 = np.array([0, 0, 0, 0])

    # Simulate the system
    def simulate_system(c_val):
        x = x0
        x1_traj = []
        for i in range(500):
            t_val = i * T / 500
            res = F(x0=x, p=vertcat(c_val, t_val))
            x = res['xf']
            x1_traj.append(x[0]-step_size)
        return sum1(vertcat(*x1_traj)**2), np.sum(np.array(x1_traj)**2)

    if check_convexity:
        # Evaluate the objective function over a range of values for c
        c_values = np.linspace(c_min, c_max, 500)
        objective_values = [simulate_system(c_val)[1] for c_val in c_values]

        # Plot the objective function
        plt.figure(figsize=(10, 6))
        plt.plot(c_values, objective_values, label='Objective Function')
        plt.xlabel('c')
        plt.ylabel('Objective Value')
        plt.title('Objective Function vs. c')
        plt.grid()

        minima = []
        for i in range(1, len(objective_values)-1):
            if objective_values[i-1] >= objective_values[i] and objective_values[i+1] >= objective_values[i]:
                minima.append(c_values[i])
                print(c_values[i])
        print(f"The function has: {'1 local minimum' if len(minima) == 1 else 'more than 1 local minima'}")

        optimal_c = minima[0]
        plt.axvline(x=minima[0], color='r', linestyle='-.', label='Optimal c')
        plt.annotate(f'{optimal_c:.3f}', xy=(optimal_c, 0.2), xytext=(optimal_c, 0.2),
            fontsize=12, color='red')
        plt.ylim(0, 1.5)
        plt.legend()
        # save the plot
        filename = f"./images/ObjectiveFunction-c.png"
        plt.savefig(filename, dpi=700, format='png', bbox_inches='tight')
        plt.show()

    # Objective function
    objective = simulate_system(c)[0]
    opti.minimize(objective)

    # Solve the optimization problem
    #opti.solver('ipopt')
    opti.solver("ipopt", {"print_time":False}, {"print_level":0}) # Suppress IPOPT output
    c_opt = opti.solve().value(c)    
    print(f"Optimal damping coefficient: {c_opt} Ns/m")

    # Simulate the system with the optimal damping coefficient
    Q2(c=c_opt, step_size=step_size)


# Run Q4_1 with different step sizes
# step_sizes = [0.01, 0.05, 0.1]
step_sizes = [0.1]
for step_size in step_sizes:
    Q4_1(step_size, check_convexity=True)
````

##### Objective Function:

\begin{aligned}
&\text{minimize} && J(K) = \int_{0}^{T} [x_1(t) - x_{1,\text{ss}}]^2 \, dt \\
&\text{subject to} && K \geq 0 \\
&&& K \leq K_{\text{max}}
\end{aligned}



where:  
$ J(K) $ is the cost function to be minimized.  
$ x_1(t) $ is the car body displacement at time \( t \).  
$ x_{1,\text{ss}} $ is the steady-state displacement of the car body.  
$ T $ is the total time of simulation.  
$ K_{\text{max}} $ is the maximum allowable gain for the skyhook control policy.

**Control Law for the Semi-active Suspension:**


$ c(t) = \begin{cases} \text{clip}\left(K \frac{\dot{x}_1}{\dot{x}_1 - \dot{x}_2}, c_{\text{min}}, c_{\text{max}}\right) & \text{if } \dot{x}_1 (\dot{x}_1 - \dot{x}_2) > 0 \\c_{\text{min}} & \text{otherwise}\end{cases} $


Where:
- $ c_{\text{min}} $ is the minimum damping coefficient,
- $ c_{\text{max}} $ is the maximum damping coefficient.


### Optimal semi-active suspension

````python
# Optimal semi-active suspension (optimal value of K)
def Q4_2(step_size, check_convexity=False):
    T = 5  # Total simulation time
    K_min = 0
    K_max = 20000

    # Define the optimization problem
    opti = Opti()

    # Decision variable and constraints (within reasonable bounds)
    K = opti.variable(1, 1)

    # Set an initial guess for K
    initial_guess = 10000
    opti.set_initial(K, initial_guess)

    # Time and state variables
    t = MX.sym('t')
    x = MX.sym('x', 4, 1)
    x1, x2, x1_dot, x2_dot = x[0], x[1], x[2], x[3]

    def casadi_semiactive_suspension(x1_dot, x2_dot, K):
        condition = x1_dot * (x1_dot - x2_dot) > 0
        c = if_else(condition, K * x1_dot / (x1_dot - x2_dot), c_min)
        c = fmin(c, c_max)
        return c

    # Equations of motion
    x1_ddot = (-k1 * (x1 - x2) - casadi_semiactive_suspension(x1_dot, x2_dot, K) * (x1_dot - x2_dot)) / m1
    x2_ddot = (-k2 * (x2 - step_size) + k1 * (x1 - x2) + casadi_semiactive_suspension(x1_dot, x2_dot, K) * (x1_dot - x2_dot)) / m2
    rhs = vertcat(x1_dot, x2_dot, x1_ddot, x2_ddot)

    # Define the ODE function
    ode = {'x': x, 'p': vertcat(K, t), 'ode': rhs}
    F = integrator('F', 'cvodes', ode, {'tf': T/1000})

    # Initial conditions
    x0 = np.array([0, 0, 0, 0])

    # Simulate the system
    def simulate_system(K_val):
        x = x0
        x1_traj_sq = []
        for i in range(1000):
            t_val = i * T / 1000
            res = F(x0=x, p=vertcat(K_val, t_val))
            x = res['xf']
            x1_traj_sq.append(x[0] - step_size)  # Append squared value
        return sum1(vertcat(*x1_traj_sq)**2), np.sum(np.array(x1_traj_sq)**2)
    
    if check_convexity:
        # Evaluate the objective function over a range of values for c
        K_values = np.linspace(K_min, K_max, 1000)
        objective_values = [simulate_system(K_val)[1] for K_val in K_values]

        # Plot the objective function
        plt.figure(figsize=(10, 6))
        plt.plot(K_values, objective_values, label='Objective Function')
        plt.xlabel('K')
        plt.ylabel('Objective Value')
        plt.title('Objective Function vs. K')
        plt.grid()
        plt.ylim(0, 0.9)

        minima = []
        for i in range(1, len(objective_values)-1):
            if objective_values[i-1] >= objective_values[i] and objective_values[i+1] >= objective_values[i]:
                minima.append(K_values[i])
                print(K_values[i])
        print(f"The function has: {'1 local minimum' if len(minima) == 1 else 'more than 1 local minima'}")

        optimal_K = minima[0]
        plt.axvline(x=optimal_K, color='r', linestyle='-.', label='Optimal K')
        plt.annotate(f'{optimal_K:.3f}', xy=(optimal_K, 0.1), xytext=(optimal_K, 0.1),
                    fontsize=12, color='red')
        plt.legend()
        # save the plot
        filename = f"./images/ObjectiveFunction-K.png"
        plt.savefig(filename, dpi=700, format='png', bbox_inches='tight')
        plt.show()

    options = {
    'ipopt': {
        'tol': 1e-10,  # Tighter optimality tolerance
        'acceptable_tol': 1e-8,  # Tighten acceptable tolerance
        }
    }

    # Objective function
    objective = simulate_system(K)[0]
    opti.minimize(objective)

    # Set the solver and options
    opti.solver('ipopt', options)
    #opti.solver("ipopt", {"print_time":False, "print_level":0}) # Suppress IPOPT output
    K_opt = opti.solve().value(K)
    print(f"Optimal gain factor: {K_opt}")

    # Simulate the system with the optimal gain factor
    Q3(K_opt, step_size)


# Run Q4_2 with different step sizes
# step_sizes = [0.01, 0.05, 0.1]
step_sizes = [0.1]
for step_size in step_sizes:
    Q4_2(step_size, check_convexity=True)
````

## Q6 - Convexity of the optimization problems (Report)
### 1. Optimal passive suspension problem: 
1. The decision variables are $c$ and $x_1(t)$. $c$ is constrained to be whithin the range $[c_{min}, c_{max}]$ which is a convex constraints.
2. The equation of motion of the car body define $\ddot{x}_1$ and $\ddot{x}_2$. These involve nonlinear terms such as products of the damping coefficient $c$ and velocity differences, making the dynamics nonlinear. (not sure about this)
3. The objective function is the squared deviation of the car body from its steady-state value. However $x_1$ depends on the nonlinear dynamics, so the relationship is nonlinear.

The optimization problem is not guaranteed to be convex because:
- The dynamics are nonlinear and introduce non-convexity into the system.
- The objective function is based on these dynamics and is thus likely to be non-convex.

### 2. Optimal semi-active suspension problem:
1. The decision variables are $K$ and $x_1(t)$. $K$ is constrained to be whithin the range $[K_{min}, K_{max}]$ which is a convex constraints.
2. The function that defines the damping coefficient $c$ is a piecewise function of the velocity differences. It uses "if_else" structure, which introduces nonlinearity due to conditional branching. Additionally, there is a min-operator adding more nonlinearity. (As a consequence it is not convex - not sure about this)
3. The equation of motion for $\ddot{x}_1$ and $\ddot{x}_2$ involve the nonlinear function "casadi_semiactive_suspension", making the dynamics non-affine and likely non-convex in $K$.

This optimization problem is not convex because:
- The function "casadi_semiactive_suspension" introduces significant nonlinearity and non-convexity.
- The dynamics depend on this nonlinear function, propagating non-convexity to the objective function.

## Q7 - Computational complexity of the optimization problems

### 1. Computational complexity analysis

The most expensive parts of the optimization problem are the computation of the Hessian of the Lagrangian and the computation of the gradient of the objective function.

- Displacement of $10 \text{ cm}$:

The computation of the gradient of the objective function is expensive because it involves evaluating how small changes in the control inputs affect the objective function. Due to the complexity of the dynamics (non-linear dynamics), this can become very time-consuming. The gradient needs to be computed at each iteration for the optimization to proceed.  

The Hessian represents the second-order derivatives, which are essential for Newton-like methods used in optimizations like IPOPT. Computing the Hessian involves a more complex and time-consuming operation since it requires calculating the curvature of the function being optimized.

### 2. Speeding up solving optimization problem
Generally, the time-consuming part is calculating derivative, to efficiently calculating the derivatives, we could use Automatic Differentiation or exploit sparsity of gradients and Hessian and symmetry for sparse Hessian when using finite difference method. But AD has already used here and FD is typically less accurate and slower compared to AD, especially in optimization problems involving stiff equations or high nonlinearity.  

So we can explore efficient computing implementation aspect. The objective function is the sum of square of the displacement difference, as mentioned in class, we can do the square part in advance and thus simplify the objective. Also, we could use parallel computing to speed up the computation.

````python
def Q4_1_optimized(step_size):
    T = 5  # Total simulation time

    # Define the optimization problem
    opti = Opti()

    # Decision variable and constraints (within reasonable bounds)
    c = opti.variable(1, 1)
    opti.subject_to(c_min <= c)
    opti.subject_to(c <= c_max)

    # Set an initial guess for c
    initial_guess = 1000
    opti.set_initial(c, initial_guess)

    # Time and state variables
    t = MX.sym('t')
    x = MX.sym('x', 4, 1)
    x1, x2, x1_dot, x2_dot = x[0], x[1], x[2], x[3]

    # Equations of motion
    x1_ddot = (-k1 * (x1 - x2) - c * (x1_dot - x2_dot)) / m1
    x2_ddot = (-k2 * (x2 - step_size) + k1 * (x1 - x2) + c * (x1_dot - x2_dot)) / m2
    rhs = vertcat(x1_dot, x2_dot, x1_ddot, x2_ddot)

    # Define the ODE function
    ode = {'x': x, 'p': vertcat(c, t), 'ode': rhs}
    F = integrator('F', 'cvodes', ode, {'tf': T/500})

    # Initial conditions
    x0 = np.array([0, 0, 0, 0])

    # Simulate the system
    def simulate_system(c_val):
        x = x0
        x1_traj_sq = []
        for i in range(500):
            t_val = i * T / 500
            res = F(x0=x, p=vertcat(c_val, t_val))
            x = res['xf']
            x_sq = x[0] * x[0]
            mid_term = 2*x[0]*step_size
            x1_traj_sq.append(x_sq - mid_term)  # Append squared value
        return sum1(vertcat(*x1_traj_sq))
    
    options = {
        'ipopt.tol': 1e-10,  # Tighter optimality tolerance
        'ipopt.acceptable_tol': 1e-8,  # Tighten acceptable tolerance
        'ipopt.hessian_approximation': 'limited-memory',
        'ipopt.jac_c_constant': 'yes',
    }

    # Objective function
    objective = simulate_system(c)
    opti.minimize(objective)

    # Set the solver and option
    opti.solver('ipopt', options)
    c_opt = opti.solve().value(c)    
    print(f"Optimal damping coefficient: {c_opt} Ns/m")

    # Simulate the system with the optimal damping coefficient
    Q2(c=c_opt, step_size=step_size)


# Run Q4_1 with different step sizes
# step_sizes = [0.01, 0.05, 0.1]
step_sizes = [0.1]
for step_size in step_sizes:
    Q4_1_optimized(step_size)
````
